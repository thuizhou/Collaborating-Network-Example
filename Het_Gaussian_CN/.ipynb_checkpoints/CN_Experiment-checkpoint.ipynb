### Example of using CN for uncertainty estimate when underlying distribution is Gaussian

```python
import sys 
import numpy as np # linear algebra
from scipy.stats import randint #to generate minibatch
import matplotlib.pyplot as plt # this is used for the plot the graph 
%matplotlib inline
from tqdm import notebook
import tensorflow as tf #tensorflow 2.0
from scipy import stats
#use interpolation for g function to solve for quantile
from scipy.interpolate import interp1d 
```

### Read data

```python
train=np.load('train.npy')
trainy=train[:,0][:,None]
trainx=train[:,1:]
test=np.load('test.npy')
testy=test[:,0][:,None]
testx=test[:,1:]
xdim=np.shape(trainx)[1]
```

### CN model

```python
def variables_from_scope(scope_name):
    """
    Returns a list of all trainable variables in a given scope. This is useful when
    you'd like to back-propagate only to weights in one part of the network
    (in our case, the generator or the discriminator).
    """
    return tf.compat.v1.get_collection(tf.compat.v1.GraphKeys.TRAINABLE_VARIABLES, scope=scope_name)
```

```python
# Graph parameters
intermediate_layer_size = 100
intermediate_layer_size2 = 60
intermediate_layer_size3 = 40
#extra term added to g-loss,small value recommended,though not bias the result
tunepara=0.05
```

```python
# f function: return the inverse cdf
def f(x,q):
    """combine x(covariate),
    percentile:q \in (0,1),
    """
    z0=tf.concat([x,q],axis=1)
    hidden_layer = tf.compat.v1.layers.dense(z0, intermediate_layer_size, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f1", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer_bn = tf.compat.v1.layers.batch_normalization(hidden_layer,name="f1bn")
    hidden_layer2 = tf.compat.v1.layers.dense(hidden_layer_bn, intermediate_layer_size2, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f2", activation=tf.keras.activations.elu,reuse=None)
    hidden_layer2_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="f2bn")
    hidden_layer3 = tf.compat.v1.layers.dense(hidden_layer2_bn, intermediate_layer_size3, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f3", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer3_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="f3bn")
    yq = tf.compat.v1.layers.dense(hidden_layer3_bn,1,kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001), name="f4", activation=None,reuse=None)
    return yq
```

```python
# g function: return the cdf Pr(Y<yq|x)
def g(yq,x):
    """combine x(covariate),
     y\in R,
     here we also match first and second moment of g
    """
    z1=tf.concat([yq,x],axis=1)
    hidden_layer = tf.compat.v1.layers.dense(z1, intermediate_layer_size,kernel_initializer=tf.compat.v1.initializers.random_normal(stddev=.001), name="g1", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer_bn = tf.compat.v1.layers.batch_normalization(hidden_layer,name="g1bn")
    hidden_layer2 = tf.compat.v1.layers.dense(hidden_layer_bn, intermediate_layer_size2, kernel_initializer=tf.compat.v1.initializers.random_normal(stddev=.001),name="g2", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer2_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="g2bn")
    gq_logit = tf.compat.v1.layers.dense(hidden_layer2_bn, 1,kernel_initializer=tf.initializers.glorot_normal, name="g3", activation=None,reuse=None)
    gq_logit_bn=tf.keras.layers.BatchNormalization(axis=-1,momentum=.5)(gq_logit)*tf.math.sqrt(3.29)
    return gq_logit_bn
```

```python
tf.compat.v1.disable_eager_execution()
tf.compat.v1.reset_default_graph()

# Placeholders
y_= tf.compat.v1.placeholder(tf.float32, [None, 1])
#used for pre-training of g
prey_= tf.compat.v1.placeholder(tf.float32, [None, 1])
x_=tf.compat.v1.placeholder(tf.float32, [None, xdim])
q_ = tf.compat.v1.placeholder(tf.float32, [None, 1])


# generate conditional quantile
with tf.compat.v1.variable_scope("f") as scope:
    yq_= f(x_,q_)

#I(Y<yq_)
ylessthan= tf.cast(tf.less_equal(y_,yq_),tf.float32)

#I(Y<prey_)
ylessthan1= tf.cast(tf.less_equal(y_,prey_),tf.float32)


with tf.compat.v1.variable_scope("g") as scope:
    # estimate Pr(Y<yq|x_)
    gq_logit = g(yq_,x_)
    gq=tf.sigmoid(gq_logit)*.9995+.0005
    scope.reuse_variables()
    # estimate Pr(Y<pre_y|x_)
    gq_logit1 = g(prey_,x_)


#Losses
f_loss = tf.reduce_mean(tf.square(q_-gq))
g_loss = 1*tf.compat.v1.losses.sigmoid_cross_entropy(ylessthan,gq_logit)+\
    tunepara*tf.compat.v1.losses.mean_squared_error(q_,gq)

#pre-loss used for pretraining for stablization 
g_loss1 = tf.compat.v1.losses.sigmoid_cross_entropy(ylessthan1,gq_logit1)

# Optimizer
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=1e-4)
pre_step= optimizer.minimize(g_loss1,var_list=variables_from_scope("g"))
g_step = optimizer.minimize(g_loss, var_list=variables_from_scope("g"))
f_step = optimizer.minimize(f_loss, var_list=variables_from_scope("f"))

# Initializer
initialize_all = tf.compat.v1.global_variables_initializer()
```

### Training: enough iteration to stablize the f and g function,check by the saved loss 

```python
 # Train the CN
sess = tf.compat.v1.Session()
sess.run(initialize_all)
gl=[]
fl=[]
batch_size = 100

iter1=50000
iter2=60000



for k in notebook.tnrange(iter1):
    i=np.random.randint(0,len(trainx),([batch_size]))
    y_tmp = trainy[i,:]
    x_tmp = trainx[i,:]
    #when we do not have f initially, we use a distribution that spans enough on real line
    prey_tmp=np.random.uniform(trainy.min()-5,trainy.max()+5,(batch_size,1))  
    sess.run(pre_step,feed_dict={y_: y_tmp,
                                x_:x_tmp,
                                prey_:prey_tmp})
    

for t in notebook.tnrange(iter2):
    i=np.random.randint(0,len(trainx),([batch_size]))
    #here we use uniform to cover different quantiles, it could be replaced by other generator,e.g beta
    #however, when the generator is no longer uniform, the g(yq,x) need to be adjusted to match the new moment
    q_tmp=np.random.uniform(0, 1.0, [batch_size, 1])
    y_tmp = trainy[i,:]
    x_tmp = trainx[i,:]
    
    for i in range(0,1):
        _,glt=sess.run([g_step,g_loss], feed_dict={y_: y_tmp,
                                    x_:x_tmp,
                                    q_:q_tmp})
    gl.append(glt)
    
    # f function has wider range(whole real line),train more iters
    for i in range(0,3):
        _,flt=sess.run([f_step,f_loss],feed_dict={y_: y_tmp,
                               x_:x_tmp,
                               q_:q_tmp})
    fl.append(flt)
 
```

```python
plt.plot(fl)
plt.plot(gl)
```

### Evaluate the calibration ($Pr(Y<y_q)=q$) with trained networks 

```python
# the percentile we are interested in evaluating calibration
per=np.linspace(0.03,0.97,20)

#calibration of CN
cncal=np.zeros(20)

# calibration of ground truth(meaningful here since it is a finite sample)
gtcal=np.zeros(20)
```

```python
for i in range(len(testx)):
    # this quantity should be carefully chosen as the study of distribution within this range
    yrange=np.linspace(testx[i,0]-40,testx[i,0]+40,25000)
    
    #numerically approximate distribution
    cdfcn=sess.run(gq ,feed_dict={x_:np.tile(testx[i,:],(25000,1)),
                                yq_:yrange[:,None]
                                }).ravel()
    
    #use iterpolation to mimic the inverse cdf from g, to generate conditional quantile
    cnquantile=interp1d(cdfcn,yrange)(per)
    
    #the ground truth
    gtquantile=stats.norm.ppf(per)*testx[i,1]+testx[i,0]
    
    #calculate calibration
    cncal=(testy[i]<cnquantile.ravel())*1.+cncal
    gtcal=(testy[i]<gtquantile.ravel())*1.+gtcal


```

```python
plt.rcParams['axes.labelsize'] = 16

plt.rcParams['axes.titlesize'] = 18

plt.rcParams['xtick.labelsize'] = 14

plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.plot(per,cncal/len(testx),'v-',label='cn')
plt.plot(per,gtcal/len(testx),'o-',label='ground truth')
plt.xlabel('Nominal Coverage of Quantile')
plt.ylabel('True Coverage of Quantile')
plt.plot([0,1],[0,1],'k--',linewidth=1,label='perfect match line')
plt.legend()
fig = plt.gcf()
fig.set_size_inches(6,6)
fig.savefig('calibration.png',transparent=True)
```

### Compare estimated cdf to ground truth on a few random samples

```python
np.random.seed(2020)
samps=np.random.randint(0,len(testx),([3]))
```

```python
samps
```

```python
mu=testx[samps,0]
sd=testx[samps,1]
```

```python
mu,sd
```

```python
yrange1=np.linspace(mu[0]-2*sd[0],mu[0]+2*sd[0],5000)
yrange2=np.linspace(mu[1]-2*sd[1],mu[1]+2*sd[1],5000)
yrange3=np.linspace(mu[2]-2*sd[2],mu[2]+2*sd[2],5000)
```

```python

cncdf1=sess.run(gq ,feed_dict={x_:np.tile(testx[samps[0],:],(5000,1)),
                                yq_:yrange1[:,None],
                                })

cncdf2=sess.run(gq ,feed_dict={x_:np.tile(testx[samps[1],:],(5000,1)),
                                yq_:yrange2[:,None],
                                })

cncdf3=sess.run(gq ,feed_dict={x_:np.tile(testx[samps[2],:],(5000,1)),
                              yq_:yrange3[:,None],
                                })
```

```python
fig = plt.gcf()
fig.set_size_inches(24,8)
plt.rcParams['axes.labelsize'] = 16

plt.rcParams['axes.titlesize'] = 18

plt.rcParams['xtick.labelsize'] = 16

plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 16
plt.subplot(1,3, 1)
plt.title(r'Sample 1 CDF, $\mu=0.018,\sigma=1.483$')
plt.plot(yrange1,cncdf1, '.',label=r'$g(.,x)$ estimated',color='orange')
plt.plot(yrange1,stats.norm.cdf((yrange1-mu[0])/sd[0]), '--',label='ground truth ',color='black')
plt.legend()


plt.subplot(1,3, 2)
plt.title(r'Sample 2 CDF, $\mu=-0.811,\sigma=1.327$')
plt.plot(yrange2,cncdf2, '.',label=r'$g(.,x)$ estimated',color='orange')
plt.plot(yrange2,stats.norm.cdf((yrange2-mu[1])/sd[1]),'--',label='ground truth',color='black')
plt.legend()



plt.subplot(1,3, 3)
plt.title(r'Sample 3 CDF, $\mu=-1.836,\sigma=1.116$')
plt.plot(yrange3,cncdf3, '.',label=r'$g(.,x)$ estimated',color='orange')
plt.plot(yrange3,stats.norm.cdf((yrange3-mu[2])/sd[2]), '--',label='ground truth ',color='black')
plt.legend()


pass

fig.savefig('cdf.png',transparent=True)
```

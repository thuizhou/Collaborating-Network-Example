## Example of using CN for uncertainty estimate when underlying distribution is Gaussian

```python
import sys 
import numpy as np # linear algebra
from scipy.stats import randint
import matplotlib.pyplot as plt # this is used for the plot the graph 
%matplotlib inline
from tqdm import notebook
import re
from tqdm import tnrange, tqdm_notebook
import seaborn as sns
import tensorflow as tf #tensorflow 2.0
from scipy import stats
from scipy import optimize
from functools import partial
```

### Generate data

```python
np.random.seed(123)
err=np.random.normal(size=(2000,1))
sig=np.random.chisquare(1,size=(2000,1))
yhat=2*np.random.normal(size=(2000,1))
y=yhat+err*sig
```

```python
yhat0=yhat[:1200,:]
yhat1=yhat[1200:,:]
sig0=sig[:1200,:]
sig1=sig[1200:,:]




#training
xhat0=np.c_[sig0,yhat0]
y0=y[:1200,:]
siz=len(y0)


#validation
xhat1=np.c_[sig1,yhat1]
y1=y[1200:,:]
siz1=len(y1)
```

### Collaborating Network

```python
def variables_from_scope(scope_name):
    """
    Returns a list of all trainable variables in a given scope. This is useful when
    you'd like to back-propagate only to weights in one part of the network
    (in our case, the generator or the discriminator).
    """
    return tf.compat.v1.get_collection(tf.compat.v1.GraphKeys.TRAINABLE_VARIABLES, scope=scope_name)
```

```python
# Graph parameters
intermediate_layer_size = 100
intermediate_layer_size2 = 50
intermediate_layer_size3 = 30
```

```python
# f function
def f(xhat,q):
    """combine xhat,
    percentile:q,
    """
    z0=tf.concat([xhat,q],axis=1)
    hidden_layer = tf.compat.v1.layers.dense(z0, intermediate_layer_size, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f1", activation=tf.compat.v1.nn.relu,reuse=None)
    hidden_layer_bn = tf.compat.v1.layers.batch_normalization(hidden_layer,name="f1bn")
    hidden_layer2 = tf.compat.v1.layers.dense(hidden_layer_bn, intermediate_layer_size2, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f2", activation=tf.keras.activations.relu,reuse=None)
    hidden_layer2_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="f2bn")
    hidden_layer3 = tf.compat.v1.layers.dense(hidden_layer2_bn, intermediate_layer_size3, kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001),name="f3", activation=tf.compat.v1.nn.relu,reuse=None)
    hidden_layer3_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="f3bn")
    yq = tf.compat.v1.layers.dense(hidden_layer3_bn,1,kernel_initializer=tf.keras.initializers.RandomNormal(stddev=.001), name="f4", activation=None,reuse=None)
    return yq
```

```python
# g function
def g(yq,xhat):
    """combine xhat,
    quantile:yq
    """
    z1=tf.concat([yq,xhat],axis=1)
    hidden_layer = tf.compat.v1.layers.dense(z1, intermediate_layer_size,kernel_initializer=tf.compat.v1.initializers.random_normal(stddev=.001), name="g1", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer_bn = tf.compat.v1.layers.batch_normalization(hidden_layer,name="g1bn")
    hidden_layer2 = tf.compat.v1.layers.dense(hidden_layer_bn, intermediate_layer_size2, kernel_initializer=tf.compat.v1.initializers.random_normal(stddev=.001),name="g2", activation=tf.compat.v1.nn.elu,reuse=None)
    hidden_layer2_bn = tf.compat.v1.layers.batch_normalization(hidden_layer2,name="g2bn")
    gq_logit = tf.compat.v1.layers.dense(hidden_layer2_bn, 1,kernel_initializer=tf.initializers.glorot_normal, name="g3", activation=None,reuse=None)
    gq_logit_bn=tf.keras.layers.BatchNormalization(axis=-1,momentum=.5)(gq_logit)*tf.math.sqrt(3.29)
    return gq_logit_bn
```

```python
tf.compat.v1.disable_eager_execution()
tf.compat.v1.reset_default_graph()
```

```python
# Placeholders
real_y = tf.compat.v1.placeholder(tf.float32, [None, 1])
pre_y= tf.compat.v1.placeholder(tf.float32, [None, 1])
xhat_=tf.compat.v1.placeholder(tf.float32, [None, 2])
q_ = tf.compat.v1.placeholder(tf.float32, [None, 1])


# f
with tf.compat.v1.variable_scope("f") as scope:
    yq_= f(xhat_,q_)


ylessthan= tf.cast(tf.less_equal(real_y,yq_),tf.float32)
ylessthan1= tf.cast(tf.less_equal(real_y,pre_y),tf.float32)


with tf.compat.v1.variable_scope("g") as scope:
    gq_logit = g(yq_,xhat_)
    gq=tf.sigmoid(gq_logit)*.9995+.0005
    scope.reuse_variables()
    gq_logit1 = g(pre_y,xhat_)


#Losses
f_loss = tf.reduce_mean(tf.square(q_-gq))
g_loss = 1*tf.compat.v1.losses.sigmoid_cross_entropy(ylessthan,gq_logit)+\
    .05*tf.compat.v1.losses.mean_squared_error(q_,gq)

#pre-loss used for pretraining for stablization 
g_loss1 = tf.compat.v1.losses.sigmoid_cross_entropy(ylessthan1,gq_logit1)

# Optimizer
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=1e-4)
pre_step= optimizer.minimize(g_loss1,var_list=variables_from_scope("g"))
g_step = optimizer.minimize(g_loss, var_list=variables_from_scope("g"))
f_step = optimizer.minimize(f_loss, var_list=variables_from_scope("f"))

# Initializer
initialize_all = tf.compat.v1.global_variables_initializer()
```

### Training: enough iteration to stablize the f and g function to be monotonic 

```python
 # Train the CN
def make_q():
    return np.random.uniform(0, 1.0, [batch_size, 1])



sess = tf.compat.v1.Session()
sess.run(initialize_all)
gl=[]
fl=[]
yltm=[]
batch_size = 150
iterations=120000



for k in notebook.tnrange(20000):
    i=np.random.randint(0,siz,([batch_size]))
    real_ytmp = y0[i,:]
    xhat_tmp=xhat0[i,:]
#when we do not have f initially, we use a uniform distribution to extract points from support
    pre_ytmp=np.random.uniform(y0.min()-2,y0.max()+2,(batch_size,1))  
    sess.run(pre_step,feed_dict={real_y: real_ytmp,
                                xhat_:xhat_tmp,
                                pre_y:pre_ytmp})
    

for t in notebook.tnrange(iterations):
#randomly generate a minibatch from valid set
    i=np.random.randint(0,siz,([batch_size]))
    qtmp=np.random.uniform(0, 1.0, [batch_size, 1])
    real_ytmp = y0[i,:]
    xhat_tmp=xhat0[i,:]
    ylessthans,gqs=sess.run([ylessthan,gq], feed_dict={real_y: real_ytmp,
                                xhat_:xhat_tmp,
                                q_:qtmp})
    yltm.append(np.mean(ylessthans))
    for i in range(0,1):
        _,glt=sess.run([g_step,g_loss], feed_dict={real_y: real_ytmp,
                                    xhat_:xhat_tmp,
                                    q_:qtmp})
    gl.append(glt)
    
    for i in range(0,3):
        _,flt=sess.run([f_step,f_loss],feed_dict={real_y: real_ytmp,
                               xhat_:xhat_tmp,
                               q_:qtmp})
    fl.append(flt)

    
#cover some tail property, optional, draw sample from beta
# for t in notebook.tnrange(10000):
#     i=np.random.randint(0,siz,([batch_size]))
#     qtmp=np.random.beta(0.5, 0.5, [batch_size, 1])
#     real_ytmp = y0[i,:]
#     xhat_tmp=xhat0[i,:]
#     ylessthans,gqs=sess.run([ylessthan,gq], feed_dict={real_y: real_ytmp,
#                                 xhat_:xhat_tmp,
#                                 q_:qtmp})
#     yltm.append(np.mean(ylessthans))
#     for i in range(0,1):
#         _,glt=sess.run([g_step,g_loss], feed_dict={real_y: real_ytmp,
#                                     xhat_:xhat_tmp,
#                                     q_:qtmp})
#     gl.append(glt)
    
#     for i in range(0,3):
#         _,flt=sess.run([f_step,f_loss],feed_dict={real_y: real_ytmp,
#                                xhat_:xhat_tmp,
#                                q_:qtmp})
#     fl.append(flt)

```

## Find quantile and evaluate the calibration

```python
#use interpolation for g function to solve for quantile
from scipy.interpolate import interp1d 
```

```python
# the percentile we are interested in studying
per=np.linspace(0.03,0.97,10)
```

```python
#calibration of our method
cncalibration=np.zeros(10)
# calibration of ground truth
thcalibration=np.zeros(10)
```

```python
ite0=np.array([])
ite1=np.array([])
ug=np.array([])
for i in range(siz1):
    xrange=np.linspace(xhat1[i,1]-40,xhat1[i,1]+40,25000)
    cdfcn=sess.run(gq ,feed_dict={xhat_:np.tile(xhat1[i,:],(25000,1)),
                                yq_:xrange[:,None]
                                }).ravel()
    cnquantile=interp1d(cdfcn,xrange)(per)
    thquantile=stats.norm.ppf(per)*xhat1[i,0]+xhat1[i,1]
    cncalibration=(y1[i]<cnquantile.ravel())*1.+cncalibration
    thcalibration=(y1[i]<thquantile.ravel())*1.+thcalibration

```

```python
plt.plot(per,cncalibration/siz1,'v-',label='cn')
plt.plot(per,thcalibration/siz1,'o-',label='ground truth model')
plt.xlabel('Predicted Level of Quantile')
plt.ylabel('True Coverage of Quantile')
plt.legend()
plt.plot([0,1],[0,1],'k--')
fig = plt.gcf()
fig.set_size_inches(10,10)
```
